```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['KaiTi'] #设置字体，能够正常显示中文
plt.rcParams['axes.unicode_minus']=False #正常显示图中负号
```

```python
fig,ax=plt.subplots(2,2,figsize=(15,12))
x=np.linspace(0,2*np.pi)
ax[0,0].plot(x,np.sin(x),c='red',linestyle='--')
ax[0,1].plot(x,np.cos(x))
ax[1,0].plot(x,np.log(x))
ax[1,1].plot(x,np.exp(x))
plt.show()
```

```python
t=np.linspace(0,2*np.pi)
x=2*np.sin(t)
y=3*np.cos(t)
plt.plot(x,y)
plt.text(0,0,r'$\frac{x^2}{2}+\frac{y^2}{3}=1$',fontsize=20)
plt.show()
```

```python
plt.subplot(1,1,1) #建立一个坐标系
x=np.arange(1,10) #指定x值
y=np.array([886,2334,5710,6482,6120,1605,3813,4428,4631])
plt.plot(x,y,color='r',linestyle='dashdot',linewidth=1,marker='o',markersize=5,label='注册用户数')
plt.title('某公司1-9月注册用户量',loc='center') #设置标题名及标题的位置

#添加数据标签
for a,b in zip(x,y):
    plt.text(a,b,b,ha='center',va='bottom',fontsize=10)
plt.grid(True) #设置网格线
plt.legend() #设置图例，调用显示出plot的label
plt.savefig(r'./images/8.1.jpg',dpi=600) #保持到本地
```

# 雷达图常用来评价某一事物，它可以直观的看出该事物的优势和不足

```python
plt.subplot(111,polar=True) #其中polar=True 表示建立一个极坐标系
dataLenth=5 #表示把整个圆分成5份
angles=np.linspace(0,2*np.pi,dataLenth,endpoint=False) #endpoint为False即不包含stop
labels=['沟通能力','业务理解能力','逻辑思维能力','快速学习能力','工具使用能力']
data=[2,3.5,4,4.5,5]
data=np.concatenate((data,[data[0]])) #闭合
angles=np.concatenate((angles,[angles[0]])) #闭合
plt.polar(angles,data,color='r',marker='o') #绘图
plt.title('某数据分析师的综合评级') 
plt.xticks(angles,labels) #设置x轴刻度
plt.savefig(r'./images/8.10.jpg',dpi=600) 
```

# 箱线图用来反应一组数据的离散情况

```python
plt.subplot(1,1,1) #建立一个坐标系
y1=np.array([886,2335,5710,6482,6120,1605,3813,4428,4631])
y2=np.array([433,1167,2855,3241,3060,802,1906,2214,2315])
x=[y1,y2] #指明x的值
labels=['注册人数','激活人数'] #指明系列类型
plt.boxplot(x,labels=labels,vert=True,widths=[0.2,0.5]) #绘图
plt.title('某公司1-9月注册与激活',loc='center')
plt.grid(False)
plt.savefig(r'./images/8.11.jpg',dpi=600) #保存到本
```

```python
plt.subplot(1,1,1) #建立一个坐标系
x=np.array([8566,5335,7310,6482])
labels=['东区','北区','南区','西区']
explode=[0.05,0,0,0] #让第一块离圆心远一点
labeldistance=1.2
plt.pie(x,labels=labels,autopct='%.0f%%',shadow=True,explode=explode,radius=1.0,labeldistance=labeldistance)
plt.title('全国各分区任务量占比',loc='center') 
plt.savefig(r'./images/8.12.jpg',dpi=600)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure()
ax=Axes3D(fig) #画三维图
x=np.arange(-4,4,0.5) #给定x值
y=np.arange(-4,4,0.5) #给定y值
x,y=np.meshgrid(x,y)
z=np.sqrt(x**2+y**2) #计算z值
ax.plot_surface(x,y,z,color='r') #绘制三维曲线图
plt.savefig(r'./images/8.16.jpg',dpi=600)
```
