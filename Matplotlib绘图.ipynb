# 图形参数设置

```python
x=np.arange(10)
y=np.arange(2,12)
plt.plot(x,y,c='red'); #绘制折线图，颜色为红色
plt.ylim(0,8); #设置 y 轴的范围 
plt.xlabel('name'); #横坐标名称
plt.ylabel('values');#纵坐标名称 
plt.xticks(range(len(x)),x); #设置坐标轴刻度
```

```python
plt.plot(x,y,linestyle='--',marker='o');
```

```python
plt.plot(x,y,'o--'); 
plt.axvline(x=1,c='red',linestyle='--'); #画垂线 x=1
plt.axhline(y=3,c='red',linestyle='--'); #做水平线 y=4
```

```python
plt.plot(x,y)
plt.text(2,7,'peak point') #在指定位置处添加文字
```

# 误差条形图

```python
s=[0.1,0.4,0.7,0.3,0.2,0.5,0.6,0.5,0.3,0.1]  #误差
plt.bar(x,y,yerr=s,error_kw={'capsize':5});
```
